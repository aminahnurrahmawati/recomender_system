Because the real datasets don't really have variation, I will try to modify the existing dataset by randomly add data to each column. 
This actions aims to avoid the overfitting that might be happening on our ML model if the datasets don't really have variation on their features. 

```python
from random import randint #for customer_id
customer_id = []
customer= 0
for i in range(150):
    customer = randint(1,100)
    customer_id.append(customer)
```

```python
page_views = []
page = 0
for i in range(150):
    page = randint(1,40)
    page_views.append(page)
```

```python

time_spent = []
time = 0
for i in range(150):
    time=randint(1,150)
    time_spent.append(time)
```

```python

product_id = []
product = 0
for i in range(150):
    product=randint(1,150)
    product_id.append(product)
```

```python
#randomize the date
import random
from datetime import datetime, timedelta

def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)

# Tanggal awal dan akhir dalam format 'YYYY-MM-DD'
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 1, 31)

purchase_date =[]
rand_date = 0
for i in range(150):
    rand_date = random_date(start_date, end_date)
    formatted_date = rand_date.strftime('%Y-%m-%d')  # date format without Hour
    purchase_date.append(formatted_date)
```

```python
category=[]
sub_category=''
num = 0
for i in range(150):
    num = randint(1,4)
    if num == 1:
        sub_category='Electronics'
    if num == 2:
        sub_category='Clothing'
    if num == 3:
        sub_category='Home & Kitchen'
    if num == 4:
        sub_category='Beauty'
    category.append(sub_category)
```

```python
price = []
sub_price=0
for i in range(150):
    sub_price = randint(10,500)
    price.append(sub_price)
```

```python
ratings =[]
rate = 0
for i in range(150):
    rate = round(random.uniform(1,5), 1) #randomize in decimal/float form
    ratings.append(rate)
```

```python
df_new = pd.DataFrame({'customer_id': customer_id, 'page_views': page_views, 'time_spent' : time_spent,	
                       'product_id' : product_id, 
                       	'purchase_date' :purchase_date,
                        'category' :category,	
                        'price': price,
                        'ratings' : ratings })
```

Below is the dataset that we obtained by randomizing the existing dataset

Furthermore I will merge the existing dataset with the new dataset

```python
dataset = pd.concat([df_merge, df_new])
```
